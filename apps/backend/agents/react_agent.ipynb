```python
import os

from dotenv import load_dotenv

load_dotenv()

from serpapi import Client as SerpApiClient

def search(query: str) -> str :
    """
    一个基于SerpApi的实战网页搜索引擎工具。
    它会智能地解析搜索结果，优先返回直接答案或知识图谱信息。
    """
    try:
        serp_api_key = os.environ.get("SERPAPI_API_KEY")
        if not serp_api_key:
            raise ValueError("未配置环境变量 SERPAPI_API_KEY")

        # 安全：避免输出敏感密钥

        client = SerpApiClient(api_key=serp_api_key)
        params = {
            "engine": "google",
            "q": query,
            "gl": "cn",
            "hl": "zh-cn",
        }
        results = client.search(params)
        # 搜索概要信息
        search_info = results.get("search_information", {})
        total_results = search_info.get("total_results", "未知")
        output_parts = [f'搜索 "{query}"，约 {total_results} 条结果：\n']

        # 知识图谱（如果有）
        kg = results.get("knowledge_graph")
        if kg:
            output_parts.append(f"📦 知识图谱：{kg.get('title', '')}")
            if kg.get("description"):
                output_parts.append(f"   {kg['description']}")
            if kg.get("website"):
                output_parts.append(f"   官网：{kg['website']}")
            output_parts.append("")

        # 相关问题（如果有）
        related = results.get("related_searches", [])
        if related:
            output_parts.append("❓ 人们还在问：")
            for q in related[:3]:
                output_parts.append(f"   - {q.get('query', '')}")
            output_parts.append("")

        # 自然搜索结果
        organic_results = results.get("organic_results", [])
        if not organic_results:
            return f'没有找到关于"{query}"的搜索结果。'

        output_parts.append("🔍 搜索结果：")
        for i, r in enumerate(organic_results[:5], 1):
            title = r.get("title", "无标题")
            link = r.get("link", "")
            snippet = r.get("snippet", "")
            date = r.get("date", "")
            date_str = f" ({date})" if date else ""
            output_parts.append(f"{i}. {title}{date_str}")
            output_parts.append(f"   链接：{link}")
            if snippet:
                output_parts.append(f"   摘要：{snippet}")
            output_parts.append("")

        return "\n".join(output_parts)
    except Exception as e:
        return f"搜索出错：{e}"

print(search("英伟达最新的GPU型号"))
```

```python
import math

_SAFE_MATH_ENV = {
    # ---------- 常量 ----------
    'pi'      : math.pi,
    'e'       : math.e,
    'tau'     : math.tau,
    'inf'     : math.inf,
    # ---------- 数学运算 ----------
    'abs'     : abs,
    'round'   : round,
    'min'     : min,
    'max'     : max,
    'sum'     : sum,
    'pow'     : pow,
    'divmod'  : divmod,
    # ---------- 三角函数 ----------
    'sin'     : math.sin,
    'cos'     : math.cos,
    'tan'     : math.tan,
    'asin'    : math.asin,
    'acos'    : math.acos,
    'atan'    : math.atan,
    'sinh'    : math.sinh,
    'cosh'    : math.cosh,
    'tanh'    : math.tanh,
    'degrees' : math.degrees,
    'radians' : math.radians,
    # ---------- 指数 / 对数 ----------
    'exp'     : math.exp,
    'log'     : math.log,
    'log10'   : math.log10,
    'log2'    : math.log2,
    'sqrt'    : math.sqrt,
    'pow2'    : lambda x: x ** 2,
    # ---------- 其他常用函数 ----------
    'ceil'      : math.ceil,
    'floor'     : math.floor,
    'factorial' : math.factorial,
    'gcd'       : math.gcd,
    'lcm'       : math.lcm,
}

def calculator(expression: str) -> str:
    """
    一个安全的高级数学计算器工具。
    支持基本四则运算、括号、幂运算、以及常见数学函数（sin/cos/tan/log/sqrt 等）。
    示例输入："(123 + 456) * 789 / 12" 或 "sqrt(144) + sin(pi/4)"
    """
    try:
        # 预处理：将各类符号统一替换为 Python 表达式可识别的形式
        _SYMBOL_MAP = str.maketrans({
            '×': '*', '÷': '/', 'x': '*', 'X': '*',
            '（': '(', '）': ')',
            '[': '(', ']': ')', '{': '(', '}': ')',
            ' ': '',  # 移除空格
        })
        expr = expression.strip().translate(_SYMBOL_MAP)
        # 安全检查：只允许数字、运算符、括号、英文点和逗号
        allowed_chars = set('0123456789+-*/().,abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ_ ')
        if not all(c in allowed_chars or c in _SAFE_MATH_ENV for c in expr):
            return f"错误：表达式包含不允许的字符。仅支持数字、运算符(+,-,*,/,**)、括号和数学函数名。"

        # 在受限环境中安全求值
        result = eval(expr, {'__builtins__': {}}, _SAFE_MATH_ENV)
        
        # 格式化输出
        if isinstance(result, float) and result.is_integer():
            return f"计算结果: {expression.strip()} = {int(result)}"
        elif isinstance(result, float):
            return f"计算结果: {expression.strip()} = {result:.10g}"
        return f"计算结果: {expression.strip()} = {result}"
    except ZeroDivisionError:
        return "错误：除数不能为零。"
    except (SyntaxError, TypeError, NameError) as e:
        return f"错误：表达式语法无效 - {e}。请检查运算符和括号是否正确配对。"
    except ValueError as e:
        return f"错误：数值范围错误 - {e}（如 sqrt 的负数参数）。"
    except OverflowError:
        return "错误：计算结果超出数值范围（溢出）。"
    except Exception as e:
        return f"计算出错：{type(e).__name__}: {e}"


# 快速测试计算器
print(calculator("(123 + 456) * 789 / 12"))
print(calculator("sqrt(144) + sin(pi/4)"))
print(calculator("2 ** 10"))
```

```python
from typing import Any, Callable

class ToolExecutor:
    """
    一个工具执行器，负责管理和执行工具。
    """
    def __init__(self):
        self.tools: dict[str, dict[str, Any]] = {}

    def registerTool(self, name: str, description: str, func: Callable[..., Any]):
        """
        向工具箱中注册一个新工具。
        """
        if name in self.tools:
            print(f"警告:工具 '{name}' 已存在，将被覆盖。")
        self.tools[name] = {"description": description, "func": func}
        print(f"工具 '{name}' 已注册。")

    def getTool(self, name: str) -> Callable[..., Any] | None:
        """
        根据名称获取一个工具的执行函数。
        """
        return self.tools.get(name, {}).get("func")

    def getAvailableTools(self) -> str:
        """
        获取所有可用工具的格式化描述字符串。
        """
        return "\n".join(
            f"- {name}: {info['description']}"
            for name, info in self.tools.items()
        )
```

```python
tool_excutor = ToolExecutor()

# 注册搜索工具
search_description = "一个网页搜索引擎。当你需要回答关于时事、事实以及在你的知识库中找不到的信息时，应使用此工具。"
tool_excutor.registerTool("Search", search_description, search)

# 注册计算器工具
calculator_description = "一个高级数学计算器。支持加减乘除、括号、幂运算(**)、以及常见数学函数(sin/cos/tan/log/sqrt/factorial等)。当用户提出数学计算问题时使用此工具，输入应为数学表达式字符串。"
tool_excutor.registerTool("Calculator", calculator_description, calculator)

print("\n--- 可用的工具 ---")
print(tool_excutor.getAvailableTools())

# 测试计算器工具
print("\n--- 执行 Action: Calculator['(123 + 456) * 789 / 12'] ---")
calc_func = tool_excutor.getTool("Calculator")
if calc_func:
    observation = calc_func("(123 + 456) * 789 / 12")
    print("--- 观察 (Observation) ---")
    print(observation)
```

```python
# ReAct 提示词模板 (使用 XML 标签格式)
REACT_PROMPT_TEMPLATE = """
你是一个有能力调用外部工具的智能助手。请严格按照 XML 格式输出你的思考过程和行动。

可用工具如下:
{tools}

输出格式要求:
- 必须使用 <thought> 和 <action> 标签
- <thought> 内写你的思考过程（可以包含多行）
- <action> 内只能包含以下两种格式之一:
    * <tool_name>参数内容</tool_name>
    * <finish>最终答案</finish>

示例:
<thought>用户问了一个数学计算题，需要调用计算器工具。</thought>
<action><Calculator>(123 + 456) * 789 / 12</Calculator></action>

<thought>已经收集到足够信息，可以给出最终答案。</thought>
<action><finish>答案是 38069.25</finish></action>

现在，请开始解决以下问题:
Question: {question}
History: {history}
"""
```

```python
import re
from typing import Any, Callable


class ReActAgent:
    def __init__(self, llm_client: GemAgentsLLM, tool_executor: ToolExecutor, max_steps: int = 5, parse_retries: int = 2):
        self.llm_client: GemAgentsLLM = llm_client
        self.tool_executor: ToolExecutor = tool_executor
        self.max_steps: int = max_steps
        self.parse_retries: int = parse_retries  # 解析失败时的最大重试次数
        self.history: list[Any] = []

    def run(self, question: str):
        """
        运行 ReAct 智能体来回答一个问题。
        解析失败时会自动重试，要求 LLM 重新输出。
        """
        self.history = []
        current_step = 0

        while current_step < self.max_steps:
            current_step += 1
            print(f"--- 第 {current_step} 步 ---")

            tools_desc = self.tool_executor.getAvailableTools()
            history_str = "\n".join(self.history)
            prompt = REACT_PROMPT_TEMPLATE.format(
                tools=tools_desc,
                question=question,
                history=history_str
            )

            response_text = self.llm_client.stream(prompt)

            # 多策略解析 + 自动重试
            thought, action, action_type = self._parse_with_retry(response_text, prompt)

            if thought:
                print(f"\n思考: {thought}")

            if action_type == "finish":
                print(f"\n最终答案: {action}")
                return action

            if action_type == "tool_call" and isinstance(action, dict):
                tool_name, tool_input = action["name"], action["input"]
                print(f"\n行动: {tool_name}[{tool_input}]")
                tool_function: Callable[..., Any] | None = self.tool_executor.getTool(tool_name)
                observation = (
                    tool_function(tool_input)
                    if tool_function
                    else f"错误：未找到名为 '{tool_name}' 的工具。"
                )
                print(f"\n观察: {observation}")
                self.history.append(f"Action: {tool_name}[{tool_input}]")
                self.history.append(f"Observation: {observation}")
            else:
                # 解析彻底失败，记录错误并继续（不终止流程）
                print("警告：解析失败，将错误信息反馈给模型。")
                self.history.append(f"Observation: 上次输出格式无法解析。请严格使用 <thought> 和 <action> 标签。原始输出：{response_text[:200]}")

        print("已达到最大步数，流程终止。")
        return None

    # ==================== 核心解析方法 ====================

    def _parse_with_retry(self, text: str, original_prompt: str) -> tuple[str | None, dict[str, Any] | str | None, str]:
        """
        带自动重试的解析入口。
        返回: (thought, action_data_or_answer, action_type)
        action_type 取值: "finish" | "tool_call" | "failed"
        """
        thought, result, atype = self._parse_output(text)
        if atype != "failed":
            return thought, result, atype

        for attempt in range(self.parse_retries):
            print(f"  [解析失败，正在重试 ({attempt + 1}/{self.parse_retries})...]")
            retry_prompt = (
                f"{original_prompt}\n\n"
                f"=== 上次输出格式有误 ===\n"
                f"你的上次输出:\n{text[:500]}\n\n"
                f"问题：未能从你的输出中解析出有效的 <action> 标签。\n"
                f"请严格按照以下格式重新输出（只输出 Thought 和 Action 两部分）:\n"
                f"<thought>你的思考</thought>\n"
                f"<action><ToolName>参数</ToolName></action>"
                f" 或 <action><finish>答案</finish></action>"
            )
            text = self.llm_client.stream(retry_prompt)
            thought, result, atype = self._parse_output(text)
            if atype != "failed":
                return thought, result, atype

        return None, None, "failed"

    def _parse_output(self, text: str) -> tuple[str | None, dict[str, Any] | str | None, str]:
        """
        多策略解析器，按优先级依次尝试:
          策略1: XML 标签格式 (<thought>/<action>)
          策略2: 旧格式 (Thought:/Action:) - 兼容性回退
          策略3: 启发式兜底（最后一行作为 Action）
        返回: (thought, action_data, action_type)
          action_type: "finish" | "tool_call" | "failed"
          finish 时 action_data 为答案字符串
          tool_call 时 action_data 为 dict {name, input}
        """
        # --- 策略1: XML 标签格式 ---
        result = self._try_parse_xml(text)
        print('************************************', result, '************************************')
        if result[2] != "failed":
            return result

        # --- 策略2: 旧 Thought:/Action: 格式 ---
        result = self._try_parse_legacy(text)
        if result[2] != "failed":
            return result

        # --- 策略3: 启发式兜底 ---
        result = self._try_parse_heuristic(text)
        return result

    # ==================== 各策略实现 ====================

    def _try_parse_xml(self, text: str) -> tuple[str | None, dict[str, Any] | str | None, str]:
        """
        策略1: 解析 XML 标签格式。
        支持: <thought>...</thought><action><ToolName>...</ToolName></action>
              <thought>...</thought><action><finish>...</finish></action>
        """
        thought_m = re.search(r"<thought\s*>([\s\S]*?)</thought\s*>", text, re.IGNORECASE)
        action_m = re.search(r"<action\s*>([\s\S]*?)</action\s*>", text, re.IGNORECASE)

        if not action_m:
            return None, None, "failed"

        thought = thought_m.group(1).strip() if thought_m else None
        action_content = action_m.group(1).strip()

        # 检测 <finish>...</finish>
        finish_m = re.match(r"<finish\s*>([\s\S]*)</finish\s*>", action_content, re.IGNORECASE)
        if finish_m:
            return thought, finish_m.group(1).strip(), "finish"

        # 检测 <ToolName>input</ToolName>
        tool_m = re.match(r"<([a-zA-Z0-9_][\w.-]{0,48})>([\s\S]*)</\1>", action_content)
        if tool_m:
            return thought, {"name": tool_m.group(1), "input": tool_m.group(2).strip()}, "tool_call"

        # action 内无有效标签，视为失败
        return None, None, "failed"

    def _try_parse_legacy(self, text: str) -> tuple[str | None, dict[str, Any] | str | None, str]:
        """
        策略2: 兼容旧的 Thought:/Action: 格式。
        支持大小写、中英文冒号、多行内容。
        """
        thought_m = re.search(
            r"(?i)Thought[：:]\s*([\s\S]*?)(?=Action[：:]|$)", text
        )
        action_m = re.search(r"(?i)Action[：:]\s*(.+)$", text, re.MULTILINE | re.DOTALL)

        if not action_m:
            return None, None, "failed"

        thought = thought_m.group(1).strip() if thought_m else None
        action_text = action_m.group(1).strip()

        # Finish[answer] → finish
        if re.match(r"(?i)Finish\s*\[", action_text):
            answer_m = re.match(r"(?i)Finish\s*\[([\s\S]*)\]$", action_text)
            if answer_m:
                return thought, answer_m.group(1).strip(), "finish"

        # ToolName[input] → tool_call
        parsed = self._parse_bracket_action(action_text)
        if parsed:
            return thought, {"name": parsed[0], "input": parsed[1]}, "tool_call"

        return None, None, "failed"

    def _try_parse_heuristic(self, text: str) -> tuple[str | None, dict[str, Any] | str | None, str]:
        """
        策略3: 启发式兜底——提取最后非空行作为 Action 尝试。
        """
        lines = [l.strip() for l in text.strip().split("\n") if l.strip()]
        if len(lines) < 1:
            return None, None, "failed"

        last_line = lines[-1]
        # 尝试作为 bracket action 解析
        parsed = self._parse_bracket_action(last_line)
        if parsed:
            thought = "\n".join(lines[:-1]) or None
            return thought, {"name": parsed[0], "input": parsed[1]}, "tool_call"

        return None, None, "failed"

    # ==================== 辅助方法 ====================

    def _parse_bracket_action(self, text: str) -> tuple[str, str] | None:
        """
        解析 ToolName[input] 格式，支持嵌套方括号。
        返回 (tool_name, tool_input) 或 None。
        """
        match = re.match(r"([a-zA-Z0-9_][\w.-]{0,48})\[(.*)", text, re.DOTALL)
        if not match:
            return None

        tool_name = match.group(1)
        bracket_start = text.find('[')
        end_pos = self._find_matching_bracket(text, bracket_start)
        if end_pos == -1:
            return None

        return (tool_name, text[bracket_start + 1 : end_pos])

    def _find_matching_bracket(self, text: str, start: int) -> int:
        """栈式查找配对的右方括号，支持嵌套。"""
        depth = 0
        for i in range(start, len(text)):
            if text[i] == '[':
                depth += 1
            elif text[i] == ']':
                depth -= 1
                if depth == 0:
                    return i
        return -1
```

```python
tool_excutor = ToolExecutor()
llm = GemAgentsLLM()

search_desc = "一个网页搜索引擎。当你需要回答关于时事、事实以及在你的知识库中找不到的信息时，应使用此工具。"
tool_excutor.registerTool(name="Search", description=search_desc, func=search)

calc_desc = "一个高级数学计算器。支持加减乘除、括号、幂运算(**)、以及常见数学函数(sin/cos/tan/log/sqrt/factorial等)。当用户提出数学计算问题时使用此工具。"
tool_excutor.registerTool(name="Calculator", description=calc_desc, func=calculator)

agent = ReActAgent(llm_client=llm, tool_executor=tool_excutor)
```
