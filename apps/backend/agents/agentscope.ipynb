```python
from agentscope.message import Msg, TextBlock

# 消息的标准结构
message = Msg(
    name="Alice",                        # 发送者名称
    content=[TextBlock(text="Hello, Bob!")],  # 消息内容（ContentBlock 列表）
    role="user",           # 角色类型
    metadata={             # 元数据信息
        "timestamp": "2024-01-15T10:30:00Z",
        "message_type": "text",
        "priority": "normal"
    }
)

print(message)
```

```python
from agentscope.message import UserMsg, AssistantMsg, SystemMsg, TextBlock, DataBlock, Base64Source

user_msg = UserMsg(name="user", content="What's in this image?")

user_msg = UserMsg(
    name="user",
    content=[
        TextBlock(text="Describe this image:"),
        DataBlock(source=Base64Source(data="...", media_type="image/png")),
    ],
)

system_msg = SystemMsg(name="system", content="You are a helpful assistant.")

assistant_msg = AssistantMsg(name="agent", content="Here is the result...")

print(user_msg)
print(system_msg)
print(assistant_msg)
```

```python
import os
from dotenv import load_dotenv
from pydantic import SecretStr

load_dotenv()

from agentscope.agent import Agent
from agentscope.model import DashScopeChatModel
from agentscope.credential import DashScopeCredential

api_key: str = os.getenv("OPENAI_API_KEY", "YOUR_API_KEY")
model: str = os.getenv("MODEL_NAME", "qwen3.7-plus") 

agent = Agent(
    name="my_agent",
    system_prompt="You are a helpful assistant.",
    model=DashScopeChatModel(
        credential=DashScopeCredential(api_key=SecretStr(api_key)),
        model=model,
    ),
)
```

```python
msg = UserMsg(name="user", content="Summarize the README in Chinese.")
async for event in agent.reply_stream(msg):
    delta = getattr(event, "delta", None)
    if delta is not None:
        print(delta, end="", flush=True)
```
